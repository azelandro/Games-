```python
def player_1_choice():
    player_1='siiii'
    while player_1 not in ['X','O']:
        player_1=input('Player 1,pick X or O: ')
        if player_1.islower():
            player_1=player_1.upper()
        if player_1 != 'X' or player_1 != 'O':
            clear_output()
            print('Incorrect letter, try again')
        if player_1=='X':
            player_2='O'
        else:
            player_2='X'
    clear_output()
    print(f'player 1 is {player_1}')
    print(f'player 2 is {player_2}')
    return player_1,player_2
```
